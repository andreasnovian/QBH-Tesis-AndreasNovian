```python
import pandas as pd
import edit_distance

filename1 = "person00001-00001-semitone-75-600.csv"
filename2 = "person00001-00001-semitone-70-200.csv"

df1 = pd.read_csv(filename1)
df2 = pd.read_csv(filename2)

ref = df1['semitone']
hyp = df2['semitone']

sm = edit_distance.SequenceMatcher(a=ref, b=hyp)
print(sm.get_opcodes())
print(sm.ratio())
print(sm.get_matching_blocks())
```
