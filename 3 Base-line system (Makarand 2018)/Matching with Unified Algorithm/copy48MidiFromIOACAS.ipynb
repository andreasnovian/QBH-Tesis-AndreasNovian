```python
import os
import shutil
import pandas as pd

folder_data = "Data query dan midi"

# dir_query = os.path.join(folder_data, "Query_IOACAS_QBH")
dir_query = os.path.join(folder_data, "Query_IOACAS_QBH_DNN-LSTM")

dir_midi = os.path.join(folder_data, "Database midi IOACAS-QBH")

# folderDestination = dir_query+"_direct to file"
queryFolderDestination = os.path.join(dir_query+"_48_MIDI")
midiFolderDestination = os.path.join(dir_midi+" 48 MIDI")
                
if not os.path.exists(queryFolderDestination):
    os.mkdir(queryFolderDestination)
if not os.path.exists(midiFolderDestination):
    os.mkdir(midiFolderDestination)

truth_file = os.path.join(dir_query, "query_truth.list")

# read csv as string type need converters
truth_df = pd.read_csv(truth_file, header=None, sep="\t", converters={i: str for i in range(100)})

midi_list = truth_df[1].values.astype(str).flatten()
temp_query_list = truth_df[0].values.astype(str).flatten()
query_list = []

for i in temp_query_list:
    query_list.append(i.replace("wavfile\\","").replace(".wav",""))

query_48_list = []
midi_48_list = []

from collections import Counter
c = Counter(midi_list)
midi_mc = c.most_common(48)

for i in midi_mc:
    midi_48_list.append(i[0])

# print(len(midi_48_list))
# print(midi_48_list)

# print(len(query_list))
# print(len(midi_list))

for i in range(len(midi_list)):
    if midi_list[i] in midi_48_list:
        query_48_list.append(query_list[i])

for i in query_48_list:
    # source = os.path.join(dir_query, i+"-parselmouth.csv")
    # destination = os.path.join(queryFolderDestination, i+"-parselmouth.csv")

    source = os.path.join(dir_query, i+"-DNN_LSTM.csv")
    destination = os.path.join(queryFolderDestination, i+"-DNN_LSTM.csv")
    
    # print(source)
    # print(destination)
    shutil.copy(source, destination)

for i in midi_48_list:
    source = os.path.join(dir_midi, i+".mid.csv")
    destination = os.path.join(midiFolderDestination, i+".mid.csv")
    
    # print(source)
    # print(destination)
    # shutil.copy(source, destination)
```
