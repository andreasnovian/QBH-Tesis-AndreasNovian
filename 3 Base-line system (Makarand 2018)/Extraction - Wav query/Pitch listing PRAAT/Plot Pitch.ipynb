```python
import pandas as pd
import plotly.express as px
```

```python
def plotGraphF0(filename):
    df = pd.read_csv(filename)
    fig = px.line(df, x = 'time', y = 'f0', title='')
    fig.show()

def plotGraphSemitone(filename):
    df = pd.read_csv(filename)
    fig = px.line(df, x = 'time', y = 'semitone', title='')
    fig.show()
```

```python
plotGraphSemitone('data/person00001-00001-semitone-70-200.csv')
```
